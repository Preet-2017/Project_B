```python
import pandas as pd

df1 = pd.read_csv("beyondb_comments_sentiment.csv")
df2 = pd.read_csv("beyondb_authors_reply_with_emotions.csv")

merged = df1.merge(df2, on=["forum_name", "thread_id"], how="left")

merged.to_csv("comments_senti_authors_reply_merged.csv", index=False)
```

```python
# Load your merged file
df = pd.read_csv("comments_senti_authors_reply_merged.csv")

# Create two dataframes based on Author_replied column
df_not_replied = df[df["Author_replied"] == 0]
df_replied = df[df["Author_replied"] == 1]

# Save them as two separate CSV files
df_not_replied.to_csv("comments_author_doesnt_reply.csv", index=False)
df_replied.to_csv("comments_author_replied.csv", index=False)

print("Files created successfully!")
```
